```python
import requests
import json
from json import JSONDecodeError
import pandas as pd
from pandas import json_normalize
import numpy as np
```

Make a .py (or .ipynb) file that contains the following (your choice of editor does not matter!) and do the following: find and access a data set in any way you want. You can use an API, a CSV, anything.

```python
url = ("https://datausa.io/api/data?drilldowns=State&measures=Population")
response = requests.get(url)
json = response.json()
state_data = pd.json_normalize(data = json, record_path = ["data"])
state_data
```

Fix TWO issues with the data set using techniques you've learned in class.

ONE: Too much info, dropping columns that are not necessary.
Cleaning to have only State, Year, and Population columns.

```python
state_data.drop(['ID State'], axis = 1, inplace=True)
state_data.drop(['ID Year'], axis = 1, inplace=True)
state_data.drop(['Slug State'], axis = 1, inplace=True)

state_data
```

TWO: Renaming 'Population' column to 'Pop.'

```python
state_data.rename(columns = {"Population":"Pop."}, inplace=True)
state_data
```
